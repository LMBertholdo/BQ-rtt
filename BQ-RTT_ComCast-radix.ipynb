### Select ALL DAYs, catchment=WAS reffering COMCAST (AS7922) (1M entries) from measurements_cluster

```python
QUERY = """
       SELECT 
          DATE (timestamp) as day,
          #catchment,
          #asn,
          src_net,
          country_ip2location,
          region,
          rtt_class,
          rtt,
          ttl
      FROM     
        `vp-measurements-testbed.datasetTestbed.measurements_cluster` 
      WHERE
         catchment IN  ('us-was-anycast01','us-mia-anycast01') 
         #AND ( (DATE (timestamp) => '2019-07-23') AND  (DATE (timestamp) <= '2019-08-02') )
         AND NOT country_ip2location = "-"
         AND NOT rtt > 1000
         AND NOT asn = "Unknown"
         AND asn = "7922"
#      GROUP BY
#        day
      ORDER BY
        day
"""
client = bigquery.Client() 
df = client.query(QUERY).to_dataframe()
df.shape
```
